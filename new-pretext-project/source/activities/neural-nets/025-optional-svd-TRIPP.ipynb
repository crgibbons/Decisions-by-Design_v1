```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics
from sklearn.preprocessing import StandardScaler
```

```python
X = np.random.randn(10, 2)
```

```python
X
```

```python
scaler = StandardScaler(with_std=True,
                        with_mean=True)
scaledX = scaler.fit_transform(X)
```

```python
scaledX
```

```python
statistics.mean(scaledX[:,1])
```

```python
U, Sig, Vt = np.linalg.svd(scaledX, full_matrices=True)
```

```python
Sig
```

```python
Vt
```

```python
V = Vt.T
```

```python
V
```

```python
scores = scaledX@V
scores
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.scatter(scaledX[:, 0], scaledX[:, 1])
ax.scatter(X[0,0], X[0, 1], color="red")
ax.arrow(0, 0, V[0, 0], V[1,0])
ax.arrow(0, 0, V[0, 1], V[1, 1])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.scatter(scores[:, 0], scores[:, 1])
ax.scatter(scores[0,0], scores[0, 1], color="red")
```
